```python
# Função para renomear os registros das perguntas
def rename_rows(row):
    if 'if she burns the food' in row:
        return 'se ela queimar a comida'
    elif 'if she argues with him' in row:
        return 'se ela discutir com ele'
    elif 'if she goes out without telling him' in row:
        return 'se ela sair sem contar a ele'
    elif 'if she neglects the children' in row:
        return 'se ela não cuida das crianças'
    elif 'if she refuses to have sex with him' in row:
        return 'se ela se recusa a fazer sexo com ele'
    elif 'for at least one specific reason' in row:
        return 'por pelo menos um motivo especifico'
    else:
        return row

# Aplicando a função para renomear as perguntas
df['question'] = df['question'].apply(rename_rows)

# Verificando as perguntas traduzidas
df['question']
```

```python
# Agrupando o dataframe por pergunta
df_questions = df.groupby('question')

# Calculando a media de 'value' (pessoas que concordam com a pergunta) para cada pergunta
average_value_per_question = df_questions['value'].mean()

# Visualizando a média por pergunta
print(average_value_per_question)
```

```python
# Criando um dataframe com as perguntas e porcentagens
new_df = pd.DataFrame({
    'question': average_value_per_question.index,
    'average_value': average_value_per_question.values
})

# Criando um gráfico de barras com cada pergunta e sua média
fig = go.Figure(data=[go.Bar(x=new_df['question'], y=new_df['average_value'])])

# Adicionando titulos
fig.update_layout(title='Média por Situação Apresentada',
                  xaxis_title='Situação',
                  yaxis_title='Média (%)')

# Mostrando o gráfico
fig.show()
```

```python
# Calculando a média geral por país
avg_country_values = df.groupby('country')['value'].mean().reset_index()

# Criando o mapa de calor
fig_map = px.choropleth(avg_country_values,
                    locations='country',
                    locationmode='country names',
                    color='value',
                    hover_name='country',
                    color_continuous_scale=px.colors.sequential.Viridis,
                    title='Média Geral por País',
                    labels={'country' : 'Páis', 'value' : 'Média (%)'})

# Personalizando o layout
fig_map.update_layout(geo=dict(showcoastlines=True))

# Exibindo o gráfico
fig_map.show()
```

```python
# Agrupando por país e pergunta, calculando média
avg_values = df.groupby(['country', 'question'])['value'].mean().reset_index()

# Criando gráfico de barras com plotly express
fig = px.bar(avg_values, x='country', y='value', color='question',
             title='Média por Situação por País',
             labels={'value': 'Media (%)', 'question': 'Situação', 'country': 'País'},
             barmode='group')

fig.show()
```

```python
# Criando um dataframe para cada grupo demografico
df_education = df[df['demographics question'] == 'Education']
df_marital = df[df['demographics question'] == 'Marital status']
df_employment = df[df['demographics question'] == 'Employment']
df_age = df[df['demographics question'] == 'Age']
df_residence = df[df['demographics question'] == 'Residence']
```

```python
# Obtendo as medias anuais para cada pergunta
question_mean_by_year = df.groupby(['question', 'survey year'])['value'].mean()

question_mean_by_year_df = question_mean_by_year.reset_index()

# Criando o gráfico de linhas
fig_year = px.line(question_mean_by_year_df, x='survey year', y='value', color='question',
              title='Médias por Ano',
              labels={'value': 'Média em (%)', 'survey year': 'Ano', 'question': 'Situação'})

pyo.plot(fig_year, filename='violencia_ano', auto_open=False)

# Exibindo o gráfico
fig_year.show()
```

```python
# Função para renomear os registros de gênero
def rename_rows(row):
    if 'M' in row:
        return 'Masculino'
    elif 'F' in row:
        return 'Feminino'
    else:
        return row

# Aplicando a função para renomear os gêneros
df_genders = df
df_genders['gender'] = df_genders['gender'].apply(rename_rows)

# Obtendo média de resposta masculina e feminina
genders = df_genders.groupby(['gender', 'question'])['value'].mean().reset_index()

# Montando o gráfico para visualizar as médias de respostas por gênero
fig_genders = px.bar(genders, x='question', y='value', color='gender',
            title='Gênero',
            barmode='group',
            labels={'value': 'Média (%)', 'survey year': 'Ano', 'question': 'Situação', 'gender' : 'Gênero'},
            category_orders={'gender' : ['Masculino', 'Feminino']})

pyo.plot(fig_genders, filename='violencia_genero', auto_open=False)

# Exibindo o gráfico
fig_genders.show()
```

```python
# Função para renomear os registros das escolaridades
def rename_rows(row):
    if 'Higher' in row:
        return 'Superior'
    elif 'Secondary' in row:
        return 'Médio'
    elif 'Primary' in row:
        return 'Fundamental'
    elif 'No education' in row:
        return 'Nenhum'
    else:
        return row

# Aplicando a função para renomear as escolaridade
df_education['demographics response'] = df_education['demographics response'].apply(rename_rows)

# Obtendo média de resposta por nivel de escolaridade
education = df_education.groupby(['demographics response', 'question'])['value'].mean().reset_index()

# Montando o gráfico para visualizar as médias de respostas por nivel de escolaridade
fig = px.bar(education, x='question', y='value', color='demographics response',
            title='Média em % de Entrevistados que Concordam com cada Situação por Nivel de Escolaridade',
            barmode='group',
            labels={'value': 'Média (%)', 'question': 'Situação', 'demographics response' : 'Escolaridade'},
            category_orders={'demographics response': ['Superior', 'Médio', 'Fundamental', 'Nenhum']})

# Exibindo o gráfico
fig.show()
```

```python
# Função para renomear os registros de relacionamento
def rename_rows(row):
    if 'Married or living together' in row:
        return 'Casado(a) ou morando junto'
    elif 'Widowed, divorced, separated' in row:
        return 'Viuvo(a), divorceado(a), separado(a)'
    elif 'Never married' in row:
        return 'Nunca se casou'
    else:
        return row

# Aplicando a função para renomear os relacionamentos
df_marital['demographics response'] = df_marital['demographics response'].apply(rename_rows)

# Obtendo média de resposta por relacionamento
marital = df_marital.groupby(['demographics response', 'question'])['value'].mean().reset_index()

# Montando o gráfico para visualizar as médias de respostas por relacionamento
fig = px.bar(marital, x='question', y='value', color='demographics response',
            title='Média em % de Entrevistados que Concordam com cada Situação por Relacionamento',
            barmode='group',
            labels={'value': 'Média (%)', 'question': 'Situação', 'demographics response' : 'Relacionamento'},
            category_orders={'demographics response': ['Casado(a) ou morando junto', 'Viuvo(a), divorceado(a), separado(a)', 'Nunca se casou']})

# Exibindo o gráfico
fig.show()
```

```python
# Função para renomear os registros de emprego
def rename_rows(row):
    if 'Employed for kind' in row:
        return 'Empregado por pagamento alternativo'
    elif 'Employed for cash' in row:
        return 'Empregado por dinheiro'
    elif 'Unemployed' in row:
        return 'Desempregado'
    else:
        return row

# Aplicando a função para renomear os empregos
df_employment['demographics response'] = df_employment['demographics response'].apply(rename_rows)

# Obtendo média de resposta por emprego
employment = df_employment.groupby(['demographics response', 'question'])['value'].mean().reset_index()

# Montando o gráfico para visualizar as médias de respostas por emprego
fig = px.bar(employment, x='question', y='value', color='demographics response',
            title='Média em % de Entrevistados que Concordam com cada Situação por Tipo de Emprego',
            barmode='group',
            labels={'value': 'Média (%)', 'question': 'Situação', 'demographics response' : 'Tipo de Emprego'},
            category_orders={'demographics response': ['Empregado por pagamento alternativo', 'Empregado por dinheiro', 'Desempregado']})

# Exibindo o gráfico
fig.show()
```

```python
# Função para renomear os registros de residencia
def rename_rows(row):
    if 'Urban' in row:
        return 'Urbana'
    else:
        return row

# Aplicando a função para renomear as residencias
df_residence['demographics response'] = df_residence['demographics response'].apply(rename_rows)

# Obtendo média de resposta por residencia
residence = df_residence.groupby(['demographics response', 'question'])['value'].mean().reset_index()

# Montando o gráfico para visualizar as médias de respostas por residencia
fig = px.bar(residence, x='question', y='value', color='demographics response',
            title='Média em % de Entrevistados que Concordam com cada Situação por Local de Residência',
            barmode='group',
            labels={'value': 'Média (%)', 'question': 'Situação', 'demographics response' : 'Residência'},
            category_orders={'demographics response': ['Urbana', 'Rural']})

# Exibindo o gráfico
fig.show()
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Criando os gráficos
fig_education = px.bar(education, x='question', y='value', color='demographics response',
                       title='Escolaridade',
                       barmode='group',
                       labels={'value': 'Média (%)', 'question': 'Situação', 'demographics response' : 'Escolaridade'},
                       category_orders={'demographics response': ['Superior', 'Médio', 'Fundamental', 'Nenhum']})
fig_marital = px.bar(marital, x='question', y='value', color='demographics response',
                     title='Estado Civil',
                     barmode='group',
                     labels={'value': 'Média (%)', 'question': 'Situação', 'demographics response' : 'Estado Civil'},
                     category_orders={'demographics response': ['Casado(a) ou morando junto', 'Viuvo(a), divorceado(a), separado(a)', 'Nunca se casou']})
fig_employment = px.bar(employment, x='question', y='value', color='demographics response',
                        title='Emprego',
                        barmode='group',
                        labels={'value': 'Média (%)', 'question': 'Situação', 'demographics response' : 'Tipo de Emprego'},
                        category_orders={'demographics response': ['Empregado por pagamento alternativo', 'Empregado por dinheiro', 'Desempregado']})
fig_age = px.bar(age, x='question', y='value', color='demographics response',
                 title='Faixa Étaria',
                 labels={'value': 'Média (%)', 'question': 'Situação', 'demographics response': 'Faixa Étaria'},
                 barmode='group',
                 hover_data={'demographics response': True})
fig_residence = px.bar(residence, x='question', y='value', color='demographics response',
                       title='Residência',
                       barmode='group',
                       labels={'value': 'Média (%)', 'question': 'Situação', 'demographics response' : 'Residência'},
                       category_orders={'demographics response': ['Urbana', 'Rural']})   # Seu gráfico de residência

# Inicializando o aplicativo Dash
app = dash.Dash(__name__)

# Layout do aplicativo
app.layout = html.Div([
    dcc.Dropdown(
        id='dropdown',
        options=[
            {'label': 'Educação', 'value': 'education'},
            {'label': 'Estado Civil', 'value': 'marital'},
            {'label': 'Emprego', 'value': 'employment'},
            {'label': 'Idade', 'value': 'age'},
            {'label': 'Residência', 'value': 'residence'}
        ],
        value='education'
    ),
    dcc.Graph(id='graph')
])

# Callback para atualizar o gráfico com base na seleção do dropdown
@app.callback(
    Output('graph', 'figure'),
    [Input('dropdown', 'value')]
)
def update_figure(selected_value):
    if selected_value == 'education':
        return fig_education
    elif selected_value == 'marital':
        return fig_marital
    elif selected_value == 'employment':
        return fig_employment
    elif selected_value == 'age':
        return fig_age
    elif selected_value == 'residence':
        return fig_residence

# Executando o aplicativo
if __name__ == '__main__':
    app.run_server(debug=True)
```
